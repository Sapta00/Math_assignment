```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def calculate_probability_distribution(distribution_name, vals):
    # Define valid distributions and their default parameters
    valid_distributions = {
        "Geometric": (stats.geom, [0.5]),  # Default probability of success p = 0.5
        "Binomial": (stats.binom, [10, 0.5]),  # Default n = 10, p = 0.5
        "Poisson": (stats.poisson, [3]),  # Default lambda = 3
        "Uniform": (stats.uniform, [0, 1])  # Default a = 0, b = 1
    }
    
    # Check if the distribution name is valid
    if distribution_name not in valid_distributions:
        return None, "Error: Invalid distribution name. Please choose from Geometric, Binomial, Poisson, or Uniform."

    # Clean vals: Remove negative values
    vals = np.array(vals)
    if np.any(vals < 0):
        vals = vals[vals >= 0]
        print("Warning: Negative values were found and removed from vals.")

    # Get the distribution and parameters
    distribution, params = valid_distributions[distribution_name]

    # Calculate probabilities
    try:
        if distribution_name == "Geometric":
            p = params[0]
            probabilities = distribution.pmf(vals, p)
        
        elif distribution_name == "Binomial":
            n, p = params
            probabilities = distribution.pmf(vals, n, p)

        elif distribution_name == "Poisson":
            lambda_param = params[0]
            probabilities = distribution.pmf(vals, lambda_param)

        elif distribution_name == "Uniform":
            a, b = params
            probabilities = distribution.pdf(vals, loc=a, scale=(b-a))

        return probabilities, None

    except Exception as e:
        return None, f"Error: {str(e)}"

def plot_distribution(distribution_name):
    # Generate x values for plotting
    if distribution_name == "Geometric":
        p = 0.5  # Default probability of success
        x = np.arange(0, 20)  # Range for geometric
        y = stats.geom.pmf(x, p)
        plt.bar(x, y, width=0.5,  color='blue')

    elif distribution_name == "Binomial":
        n, p = 10, 0.5  # Default n and p
        x = np.arange(0, n + 1)
        y = stats.binom.pmf(x, n, p)
        plt.bar(x, y, width=0.5,  color='blue')

    elif distribution_name == "Poisson":
        lambda_param = 3  # Default lambda
        x = np.arange(0, 20)  # Range for Poisson
        y = stats.poisson.pmf(x, lambda_param)
        plt.bar(x, y, width=0.5, color='blue')

    elif distribution_name == "Uniform":
        a, b = 0, 1  # Default a and b
        x = np.linspace(a, b, 100)
        y = stats.uniform.pdf(x, loc=a, scale=(b-a))
        plt.plot(x, y, color='blue')

    plt.title(f"{distribution_name} Distribution")
    plt.xlabel("x")
    plt.ylabel("Probability")
    plt.legend()
    plt.grid()
    plt.show()
 # Example usage
if __name__ == "__main__":
    # User input for distribution name
    distribution_name = input("Enter the name of the distribution (Geometric, Binomial, Poisson, Uniform): ")
    
    # Input values of x
    vals = list(map(float, input("Enter the values of x (comma-separated): ").split(',')))

    # Calculate probabilities
    probabilities, error_message = calculate_probability_distribution(distribution_name, vals)
    if error_message:
        print(error_message)
    else:
        print("Calculated probabilities:", probabilities)

        # Plot the distribution for the specified distribution name
        plot_distribution(distribution_name)

        # Plot all distributions one after the other
        for dist in ["Binomial", "Poisson", "Uniform"]:
            print(f"\nPlotting {dist} Distribution:")
            plot_distribution(dist)
```
